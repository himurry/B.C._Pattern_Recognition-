**3. En el modelado estadístico, el análisis de regresión es un proceso para estimar la relación entre variables. Investigue y describa la regresión logística.**


La regresión logística es un proceso de modelado de la probabilidad de un resultado discreto dada una variable de entrada; se utiliza cuando la variable dependiente (objetivo) es categórica. La regresión logística se utiliza para describir datos y explicar la relación entre una variable binaria dependiente y una o más variables independientes nominales, ordinales, de intervalo o de razón. El modelo más común de regresión logística es el binario, aunque también existe el modelo multinomial, que se utiliza cuando hay más de dos posibles resultados discretos en el que no importa el orden, y el modelo ordinal, en el que hay más de dos posibles resultados discretos y sí importa el orden.
La regresión logística utiliza la función sigmoidea. El eje vertical representa la probabilidad de una clasificación dada y el eje horizontal es el valor de x. Se asume que la distribución de y|x es la distribución de Bernoulli.



**4. Clasifique los datos mediante regresión logística y mida el desempeño de su modelo. Describa la medida usada para medir el desempeño de su modelo.**

Importando bibliotecas para analítica

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_shuffle = df_iris.sample(frac = 1) 
df_shuffle = df_shuffle.sample(frac = 1) 

x = np.array(df_shuffle.drop(['class'], 1)) 
y = np.array(df_shuffle['class']) 
```

Separación de nuestra muestra y entrenamiento del modelo 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.75) 
lr_model = LogisticRegression() 
lr_model.fit(x_train, y_train) 
```

```python
Y_pred = lr_model.predict(x_test) 
print('Precisión: ',lr_model.score(x_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, Y_pred)
print(conf_matrix)
```

```python
cm = pd.DataFrame(
    conf_matrix, index = [i for i in types],
    columns = [i for i in types]
)
plt.figure(figsize = (12,7))
sns.heatmap(cm, annot=True, fmt="d")
plt.show()
```
